```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_theme(style="darkgrid", palette="pastel")
```

```python
df = pd.read_csv("../data/diabetes_cleaned.csv")

df["diabetes"].value_counts().plot(kind="bar")
plt.title("Diabetes Distribution")
plt.xticks([0,1], ["No Diabetes", "Diabetes"], rotation=0)
plt.show()
```

Approximately 8.5 % of individuals in the dataset have diabetes, while 91.5 % do not.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="diabetes",
    y="HbA1c_level",
    data=df
)

plt.title("HbA1c Levels by Diabetes Status", fontsize=14)
plt.xlabel("Diabetes", fontsize=12)
plt.ylabel("HbA1c Level", fontsize=12)

plt.xticks([0,1], ["No Diabetes", "Diabetes"])

plt.show()
```

## HbA1c and diabetes

The visualization shows that individuals with diabetes have significantly higher HbA1c levels compared to those without diabetes. This result aligns with medical expectations and supports the validity of the dataset.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="diabetes",
    y="blood_glucose_mmol",
    data=df
)

plt.title("Blood Glucose (mmol/l) by Diabetes Status")
plt.xlabel("Diabetes")
plt.ylabel("Blood Glucose (mmol/l)")

plt.xticks([0,1], ["No Diabetes", "Diabetes"])

sns.despine()

plt.show()
```

Individuals with diabetes show clearly higher blood glucose levels compared to non-diabetic individuals.

Normal fasting blood glucose levels are typically around 4-6 mmol/l. However, the values in this dataset appear to represent non-fasting (random) blood glucose measurements. 
This should be considered when interpreting the results.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="diabetes",
    y="bmi",
    data=df
)

plt.title("BMI by Diabetes Status")
plt.xlabel("Diabetes")
plt.ylabel("BMI")

plt.xticks([0,1], ["No Diabetes", "Diabetes"])

sns.despine()

plt.show()
```

## BMI and diabetes

Individuals with diabetes tend to have higher BMI values compared to non-diabetic individuals.
This suggests a potential association between higher body weight and increased risk of diabetes.

```python
hypertension_percent = df.groupby("hypertension")["diabetes"].mean() * 100

plt.figure(figsize=(6,4))

ax = hypertension_percent.plot(kind="bar")

plt.title("Percentage of Individuals with Diabetes")
plt.xlabel("Hypertension Status")
plt.ylabel("Diabetes (%)")

plt.xticks([0,1], ["No Hypertension", "Hypertension"], rotation=0)

# lisätään arvot pylväisiin
for i, v in enumerate(hypertension_percent):
    plt.text(i, v + 0.5, f"{v:.1f}%", ha='center')

sns.despine()

plt.show()
```

## Hypertension and diabetes

Among individuals without hypertension, approximately 6.9 % have diabetes.
In contrast, 27.9 % of individuals with hypertension have diabetes.

This suggests a strong association between hypertension and diabetes.
